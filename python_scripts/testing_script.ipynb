```python
#Final text Clean up script
test_df['text'] = test_df['text'].str.replace("\\", "")

test_df['text'] = test_df['text'].str.replace("*", "")
test_df['text'] = test_df['text'].str.replace(".", "")
test_df['text'] = test_df['text'].str.replace("!", "")
test_df['text'] = test_df['text'].str.replace("?", "")
test_df['text'] = test_df['text'].str.replace(",", "")
test_df['text'] = test_df['text'].str.replace("-", " ")
test_df['text'] = test_df['text'].str.replace('"', "")
test_df['text'] = test_df['text'].str.replace("'ve'", " have")
test_df.loc['text'] = test_df['text'].str.replace("'ll", " will")
test_df['text'] = test_df['text'].str.replace("(", "")
test_df['text'] = test_df['text'].str.replace(")", "")
test_df['text'] = test_df['text'].str.replace("'", "")
test_df['text'] = test_df['text'].str.replace("/", "")
test_df['text'] = test_df['text'].str.lower()
```
